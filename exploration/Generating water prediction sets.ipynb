```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

pd.options.display.max_columns = None
```

```python
%%time
result = pd.read_csv('../data/water/CA-result-withmeasuregroup.csv', low_memory=False)
result = result.ix[result.StartDate >= '2010-01-01']
```

```python
%%time
result.ix[result.WarningLevel == 'Green', 'WarningCode'] = 0
result.ix[result.WarningLevel == 'Amber', 'WarningCode'] = 1
result.ix[result.WarningLevel == 'Red', 'WarningCode'] = 2
```

```python
%%time
dailyMaximums = result[['LocationIdentifier', 'StartDate', 'Pollutant', 'WarningCode']]\
                      .groupby(['LocationIdentifier', 'StartDate', 'Pollutant'])\
                      .max()\
                      .reset_index()
```

```python
dailyMaximums.columns
```

```python
result.columns
```

```python
station = pd.read_csv('../data/water/CA-station-clean.csv', low_memory=False,
                     usecols=['MonitoringLocationId', 'Latitude', 'Longitude'])
data = pd.merge(dailyMaximums, station, left_on='LocationIdentifier', right_on='MonitoringLocationId')
data = data[['LocationIdentifier', 'Latitude', 'Longitude', 'StartDate', 'Pollutant', 'WarningCode']]
data.ix[data.WarningCode == 0, 'WarningLevel'] = 'Green'
data.ix[data.WarningCode == 1, 'WarningLevel'] = 'Amber'
data.ix[data.WarningCode == 2, 'WarningLevel'] = 'Red'
data
```

```python
data.to_csv('../data/water/prediction/warning_level_data.csv', index=False)
```
